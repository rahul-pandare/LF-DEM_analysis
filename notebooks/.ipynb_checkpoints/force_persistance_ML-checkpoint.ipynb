```python
import numpy as np
import glob

## Reading .npy files
topDir      = '/Users/rahul/Downloads/phi_0.78'
sigma       = [10, 20, 50]
persistance = {}
viscosity   = {}

clip_value = 1e-4

# start = 200
# end   = start + 200
frames = 200

for stress in sigma:
    file_list = sorted(glob.glob(f"{topDir}/sigma_{stress}/numpy_arrays/*.npy"))  # sort to keep consistent order

    arrays              = [np.load(f) for f in file_list]
    persistance[stress] = np.stack(arrays, axis=0)

    data = np.loadtxt(glob.glob(f"{topDir}/sigma_{stress}/data_*.dat")[0])
    viscosity[stress] = list(data[-frames-1:, 3])

print(persistance[sigma[1]].shape) #(201, 100, 100)
print(len(viscosity[sigma[1]])) # 201

pers_min =  np.inf
pers_max = -np.inf
visc_min =  np.inf
visc_max = -np.inf

for key in persistance.keys():
    pers     = persistance[key]
    pers_min = min(pers_min, pers.min())
    pers_max = max(pers_max, pers.max())
    
    visc     = viscosity[key]
    visc_min = min(visc_min, np.min(visc))
    visc_max = max(visc_max, np.max(visc))

for key in persistance.keys():
    pers = persistance[key]
    visc = viscosity[key]
    persistance[key] = np.clip(((pers - pers_min) / (pers_max - pers_min)) , clip_value, None)
    viscosity[key]   = [(i - visc_min)/ (visc_max - visc_min) for i in visc]
    
for key in persistance.keys():
    pers     = persistance[key]
    pers_min = pers.min()
    pers_max = pers.max()
    
    visc     = viscosity[key]
    visc_min = np.min(visc)
    visc_max = np.max(visc)
    
    print('\nnormalised')
    print("Persistence min:", pers_min)
    print("Persistence max:", pers_max)
    print("Viscosity min:", visc_min)
    print("Viscosity max:", visc_max)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# -------------------------------
# Custom Dataset
# -------------------------------
class PersistenceDataset(Dataset):
    def __init__(self, X, y):
        """
        X: np.array, shape (N, H, W)
        y: np.array or list, shape (N,)
        """
        self.X = torch.tensor(X, dtype=torch.float32).unsqueeze(1)  # add channel dimension
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# -------------------------------
# CNN Model for Regression
# -------------------------------
class CNNRegressor(nn.Module):
    def __init__(self):
        super(CNNRegressor, self).__init__()
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4))  # shrink to fixed size
        )

        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 4 * 4, 128),
            nn.ReLU(),
            nn.Linear(128, 1)  # regression output
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x.squeeze(1)

class ImprovedCNNRegressor(nn.Module):
    def __init__(self):
        super(ImprovedCNNRegressor, self).__init__()
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1)),
        )

        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 1)
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x.squeeze(1)

# -------------------------------
# Training Function
# -------------------------------
def train_model(persistence, viscosity, epochs=30, batch_size=32, lr=1e-3, test_size=0.2):
    # Combine all sigma datasets
    X = []
    y = []
    for sigma in persistence.keys():
        X.append(persistence[sigma])
        y.append(viscosity[sigma])
    X = np.concatenate(X, axis=0)  # (N_total, H, W)
    y = np.concatenate(y, axis=0)  # (N_total,)

    # Train/test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, shuffle=True, random_state=42
    )

    # Create datasets
    train_ds = PersistenceDataset(X_train, y_train)
    test_ds  = PersistenceDataset(X_test, y_test)

    # Create loaders
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
    test_loader  = DataLoader(test_ds, batch_size=batch_size)

    # Initialize model, loss, optimizer
    model = CNNRegressor()                ## Change here
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    # Training loop
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        for Xb, yb in train_loader:
            optimizer.zero_grad()
            preds = model(Xb)
            loss  = criterion(preds, yb)
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * Xb.size(0)

        # Validation on test set
        model.eval()
        test_loss = 0
        with torch.no_grad():
            for Xb, yb in test_loader:
                preds = model(Xb)
                loss  = criterion(preds, yb)
                test_loss += loss.item() * Xb.size(0)

        print(f"Epoch {epoch+1}/{epochs} "
              f"Train Loss: {train_loss/len(train_ds):.4f} "
              f"Test Loss: {test_loss/len(test_ds):.4f}")

    return model

# -------------------------------
# Example Usage
# -------------------------------
model = train_model(persistance, viscosity, epochs=100, batch_size=32)
```

```python
import torch

def evaluate_model(model, test_loader):
    model.eval()  # set model to evaluation mode
    y_true = []
    y_pred = []

    with torch.no_grad():
        for Xb, yb in test_loader:
            preds = model(Xb)
            y_true.append(yb.numpy())
            y_pred.append(preds.numpy())

    # Concatenate all batches
    y_true = np.concatenate(y_true)
    y_pred = np.concatenate(y_pred)

    return y_true, y_pred

# Example usage
# Train/test split
X = []
y = []
for sigma in persistance.keys():
    X.append(persistance[sigma])
    y.append(viscosity[sigma])
X = np.concatenate(X, axis=0)  # (N_total, H, W)
y = np.concatenate(y, axis=0)  # (N_total,)

test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, shuffle=True, random_state=42
    )

test_ds     = PersistenceDataset(X_test, y_test)  # use your test split
test_loader = DataLoader(test_ds, batch_size=32)
y_true, y_pred = evaluate_model(model, test_loader)
```

```python
intFile = '/Volumes/rahul_2TB/high_bidispersity/new_data/NP_1000/phi_0.75/ar_1.4/Vr_0.5/run_1/int_random_seed_params_stress100r_shear.dat'
intlist = readFiles.interactionsList(open(intFile))

data = np.loadtxt(open('/Volumes/rahul_2TB/high_bidispersity/new_data/NP_1000/phi_0.75/ar_1.4/Vr_0.5/run_1/data_random_seed_params_stress100r_shear.dat'))
```

```python
off = 100
normAll = []
tangAll = []
visc    = list(data[off:, 3])

for frame in intlist[off:]:
    lubNormForce = frame[:, 6]
    lubTangForce = np.linalg.norm(frame[:, [7, 8, 9]], axis=1)
    
    contNormForce = frame[:, 11]
    contTangForce = np.linalg.norm(frame[:, [12, 13, 14]], axis=1)
    
    totalNorm = lubNormForce + contNormForce
    totalTang = lubTangForce + contTangForce
    
    normAll.append(np.sum(totalNorm))
    tangAll.append(np.sum(totalTang))
```

```python
#plt.scatter(visc, np.add(normAll, tangAll))
plt.scatter(visc, normAll)
# plt.xscale("log")
# plt.yscale("log")
plt.show()
```

```python
## Dimentional analysis

plt.scatter(visc, normAll, label="Data", alpha=0.7)

# Linear fit (y = m*x + c)
coeffs = np.polyfit(visc, normAll, 1)  # degree=1 for linear
m, c = coeffs
fit_line = np.poly1d(coeffs)

# Generate smooth x for line
x_fit = np.linspace(min(visc), max(visc), 200)
y_fit = fit_line(x_fit)

# Residuals and standard deviation
residuals = normAll - fit_line(visc)
std = np.std(residuals)

# Plot fit line
plt.plot(x_fit, y_fit, color="red",
         label=fr"$y = {m:.2f}x \;+\; {c:.2f}$")

# Add error envelope (±1 std dev)
plt.fill_between(x_fit, y_fit - std, y_fit + std,
                 color="red", alpha=0.2, label=r"$\pm 1\sigma$")

plt.xlabel(r"$\eta_r$", fontsize= 13)
plt.ylabel(r"$\sum \mathbf{F^{n}_{lub, cont}}$", fontsize =13)
plt.legend()
plt.show()
```
