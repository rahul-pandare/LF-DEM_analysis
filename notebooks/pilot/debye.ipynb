```python
## flow curve (for new data)

#3D (comparing with sids data)

npp = 500
vr  = '0.5'
phi = 0.5
ar  = 1.4
off = 100
numRun   = 1
stress   = ['0.01', '0.05', '0.1', '0.12', '0.15', '0.18', '0.2', '0.5', 
            '1', '2','5', '10', '25', '50', '80', '100']#, '200', '250']
stress1  = [float(s) for s in stress]
debyelen = ['0.1', '0.05']
markers  = ['o', '*', 'p', 's']
colors   = ['#4a91b5', '#e68139', '#5da258', '#87629b']

#linux
topDir        = '/home/rahul/mnt/sshfs_expanse/simulations/debye/3D/'
fig_save_path = '/home/rahul/City College Dropbox/Rahul Pandare/CUNY/research/bidisperse_project/figures/debye/'

#mac
#topDir = "/Volumes/rahul_2TB/high_bidispersity/new_data/flow_curves"
#fig_save_path = '/Users/rahul/City College Dropbox/Rahul Pandare/CUNY/research/bidisperse_project/conferences/GSOE_poster/figures/'

fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # correct syntax

for j, dlj in enumerate(debyelen):
    visc     = [np.nan for _ in stress1]
    srate    = [np.nan for _ in stress1]
    visc_std = [np.nan for _ in stress1]

    for s in range(len(stress)):
        workDir = f'{topDir}{dlj}as/stress_{stress[s]}r/NP_{npp}/phi_{phi:.2f}/ar_{ar}/Vr_{vr}/run_{numRun}'
        if os.path.exists(workDir):
            file_name = glob.glob(f'{workDir}/data_*.dat')[0]
            data_array = np.loadtxt(file_name)

            if data_array.ndim != 1:
                strain = data_array[:, 1]
                data   = data_array[:, 3][off:]  # viscosity
                rate   = data_array[:, 2][off:]  # shear rate

                if len(data) > 0 and all(data >= 0):
                    visc[s]     = np.mean(data)
                    srate[s]    = np.mean(rate)
                    visc_std[s] = np.std(data)

    visc_plot     = [i for i in visc if not np.isnan(i)]
    srate_plot    = [i for i in srate if not np.isnan(i)]
    visc_err_plot = [i for i in visc_std if not np.isnan(i)]
    stress_plot   = [stress1[i] for i in range(len(stress1)) if not np.isnan(visc[i])]

    # viscosity subplot
    ax[0].plot(
        stress_plot, visc_plot,
        linestyle='--', marker=markers[j],
        label=fr'$\lambda = {dlj} a_{{s}}$',
        color=colors[j], linewidth=1.7, markersize=6.5)

    # shear rate subplot
    ax[1].plot(
        srate_plot, visc_plot,
        linestyle='--', marker=markers[j],
        label=fr'$\lambda = {dlj} a_{{s}}$',
        color=colors[j], linewidth=1.7, markersize=6.5)


for axi in ax:
    axi.spines['top'].set_linewidth(1)
    axi.spines['right'].set_linewidth(1)
    axi.spines['left'].set_linewidth(1)
    axi.spines['bottom'].set_linewidth(1)

    axi.grid(which='both', alpha=0.25)
    axi.set_xscale('log')
    #axi.set_yscale('log')

ax[0].set_xlabel(r'$\sigma / \sigma_{0}$', fontsize=16, fontweight='bold')
ax[0].set_ylabel(r'$\eta_{r}$', fontsize=16, fontweight='bold')
#ax[0].set_title("Relative Viscosity", fontsize=16, pad=10)

ax[1].set_xlabel(r'$\dot{\gamma}$', fontsize=16, fontweight='bold')
ax[1].set_ylabel(r'$\eta_{r}$', fontsize=16, fontweight='bold')
#ax[1].set_title("Shear Rate", fontsize=16, pad=10)

fig.suptitle(fr'$N= {npp}, \;\phi = {phi}, \; \delta = {ar}, \; \zeta = {vr}, \; \mu =1$',fontsize=18, y=0.92, fontweight='bold')
ax[0].legend(loc='lower right', fontsize=13, framealpha=0.5)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

if True:
    figFormat = ".pdf"
    plt.savefig(fig_save_path + "flowcurve_debye_stress_3D" + figFormat, bbox_inches="tight", dpi=600)

plt.show()
```
